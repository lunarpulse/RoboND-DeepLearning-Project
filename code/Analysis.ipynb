```python
with open('../writeup/data/loss_0_2_3.txt', 'r') as f:
    pattern_loss = r"(\d\.\d+)\W\-\W[a-z_]+\:\W(\d\.\d+)"
    re_loss = re.compile(pattern_loss)
    #print(f.read())
    f.seek(0)
    data_loss= re_loss.findall(f.read())
    #print(data_loss)
    pattern_time = r"-\W(\d+)s"    
    re_time = re.compile(pattern_time)       
    f.seek(0)
    data_time= re_time.findall(f.read())
    #print(data_time)
    prev_loss = 0
    prev_time = 0
    accumulated_seconds = 0
    time_data = []
    for index, elasped in enumerate(data_time):
        accumulated_seconds += int(elasped)
        derivative_train_loss = (float(data_loss[index][0]) - prev_loss)/float(elasped)
        derivative_val_loss = (float(data_loss[index][1]) - prev_loss)/float(elasped)

        #print(elasped, accumulated_seconds)
        time_data.append((int(elasped), accumulated_seconds, derivative_train_loss,derivative_val_loss))
        
    out_data = [data_loss[i]+t_data for i, t_data in enumerate(time_data)]
    #print(out_data)
    with open('../writeup/data/analysis0_2_3.csv','w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['Loss','ValLoss','time per step','accumulated seconds', 'derivative_train_loss','derivative_val_loss'])
        for row in out_data:
            csv_out.writerow(row)
            
with open('../writeup/data/loss_0_2_4.txt', 'r') as f:
    pattern_loss = r"(\d\.\d+)\W\-\W[a-z_]+\:\W(\d\.\d+)"
    re_loss = re.compile(pattern_loss)
    #print(f.read())
    f.seek(0)
    data_loss= re_loss.findall(f.read())
    #print(data_loss)
    pattern_time = r"-\W(\d+)s"    
    re_time = re.compile(pattern_time)       
    f.seek(0)
    data_time= re_time.findall(f.read())
    #print(data_time)
    prev_loss = 0
    prev_time = 0
    accumulated_seconds = 0
    time_data = []
    for index, elasped in enumerate(data_time):
        accumulated_seconds += int(elasped)
        derivative_train_loss = (float(data_loss[index][0]) - prev_loss)/float(elasped)
        derivative_val_loss = (float(data_loss[index][1]) - prev_loss)/float(elasped)

        #print(elasped, accumulated_seconds)
        time_data.append((int(elasped), accumulated_seconds, derivative_train_loss,derivative_val_loss))
        
    out_data = [data_loss[i]+t_data for i, t_data in enumerate(time_data)]
    #print(out_data)
    with open('../writeup/data/analysis0_2_4.csv','w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['Loss','ValLoss','time per step','accumulated seconds', 'derivative_train_loss','derivative_val_loss'])
        for row in out_data:
            csv_out.writerow(row)

with open('../writeup/data/loss_0_2_5.txt', 'r') as f:
    pattern_loss = r"(\d\.\d+)\W\-\W[a-z_]+\:\W(\d\.\d+)"
    re_loss = re.compile(pattern_loss)
    #print(f.read())
    f.seek(0)
    data_loss= re_loss.findall(f.read())
    #print(data_loss)
    pattern_time = r"-\W(\d+)s"    
    re_time = re.compile(pattern_time)       
    f.seek(0)
    data_time= re_time.findall(f.read())
    #print(data_time)
    prev_loss = 0
    prev_time = 0
    accumulated_seconds = 0
    time_data = []
    for index, elasped in enumerate(data_time):
        accumulated_seconds += int(elasped)
        derivative_train_loss = (float(data_loss[index][0]) - prev_loss)/float(elasped)
        derivative_val_loss = (float(data_loss[index][1]) - prev_loss)/float(elasped)

        #print(elasped, accumulated_seconds)
        time_data.append((int(elasped), accumulated_seconds, derivative_train_loss,derivative_val_loss))
        
    out_data = [data_loss[i]+t_data for i, t_data in enumerate(time_data)]
    #print(out_data)
    with open('../writeup/data/analysis0_2_5.csv','w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['Loss','ValLoss','time per step','accumulated seconds', 'derivative_train_loss','derivative_val_loss'])
        for row in out_data:
            csv_out.writerow(row)
            
with open('../writeup/data/loss_0_2_6.txt', 'r') as f:
    pattern_loss = r"(\d\.\d+)\W\-\W[a-z_]+\:\W(\d\.\d+)"
    re_loss = re.compile(pattern_loss)
    #print(f.read())
    f.seek(0)
    data_loss= re_loss.findall(f.read())
    #print(data_loss)
    pattern_time = r"-\W(\d+)s"    
    re_time = re.compile(pattern_time)       
    f.seek(0)
    data_time= re_time.findall(f.read())
    #print(data_time)
    prev_loss = 0
    prev_time = 0
    accumulated_seconds = 0
    time_data = []
    for index, elasped in enumerate(data_time):
        accumulated_seconds += int(elasped)
        derivative_train_loss = (float(data_loss[index][0]) - prev_loss)/float(elasped)
        derivative_val_loss = (float(data_loss[index][1]) - prev_loss)/float(elasped)

        #print(elasped, accumulated_seconds)
        time_data.append((int(elasped), accumulated_seconds, derivative_train_loss,derivative_val_loss))
        
    out_data = [data_loss[i]+t_data for i, t_data in enumerate(time_data)]
    #print(out_data)
    with open('../writeup/data/analysis0_2_6.csv','w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['Loss','ValLoss','time per step','accumulated seconds', 'derivative_train_loss','derivative_val_loss'])
        for row in out_data:
            csv_out.writerow(row)
            
with open('../writeup/data/loss_3.txt', 'r') as f:
    pattern_loss = r"(\d\.\d+)\W\-\W[a-z_]+\:\W(\d\.\d+)"
    re_loss = re.compile(pattern_loss)
    #print(f.read())
    f.seek(0)
    data_loss= re_loss.findall(f.read())
    #print(data_loss)
    pattern_time = r"-\W(\d+)s"    
    re_time = re.compile(pattern_time)       
    f.seek(0)
    data_time= re_time.findall(f.read())
    #print(data_time)
    prev_loss = 0
    prev_time = 0
    accumulated_seconds = 0
    time_data = []
    for index, elasped in enumerate(data_time):
        accumulated_seconds += int(elasped)
        derivative_train_loss = (float(data_loss[index][0]) - prev_loss)/float(elasped)
        derivative_val_loss = (float(data_loss[index][1]) - prev_loss)/float(elasped)

        #print(elasped, accumulated_seconds)
        time_data.append((int(elasped), accumulated_seconds, derivative_train_loss,derivative_val_loss))
        
    out_data = [data_loss[i]+t_data for i, t_data in enumerate(time_data)]
    #print(out_data)
    with open('../writeup/data/analysis_3.csv','w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['Loss','ValLoss','time per step','accumulated seconds', 'derivative_train_loss','derivative_val_loss'])
        for row in out_data:
            csv_out.writerow(row)
            
with open('../writeup/data/loss_4.txt', 'r') as f:
    pattern_loss = r"(\d\.\d+)\W\-\W[a-z_]+\:\W(\d\.\d+)"
    re_loss = re.compile(pattern_loss)
    #print(f.read())
    f.seek(0)
    data_loss= re_loss.findall(f.read())
    #print(data_loss)
    pattern_time = r"-\W(\d+)s"    
    re_time = re.compile(pattern_time)       
    f.seek(0)
    data_time= re_time.findall(f.read())
    #print(data_time)
    prev_loss = 0
    prev_time = 0
    accumulated_seconds = 0
    time_data = []
    for index, elasped in enumerate(data_time):
        accumulated_seconds += int(elasped)
        derivative_train_loss = (float(data_loss[index][0]) - prev_loss)/float(elasped)
        derivative_val_loss = (float(data_loss[index][1]) - prev_loss)/float(elasped)

        #print(elasped, accumulated_seconds)
        time_data.append((int(elasped), accumulated_seconds, derivative_train_loss,derivative_val_loss))
        
    out_data = [data_loss[i]+t_data for i, t_data in enumerate(time_data)]
    #print(out_data)
    with open('../writeup/data/analysis_4.csv','w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['Loss','ValLoss','time per step','accumulated seconds', 'derivative_train_loss','derivative_val_loss'])
        for row in out_data:
            csv_out.writerow(row)
```
